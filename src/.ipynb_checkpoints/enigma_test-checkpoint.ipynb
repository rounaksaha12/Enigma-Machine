```python
import numpy as np
import pandas as pd
```

```python
alpha_lower='abcdefghijklmnopqrstuvwxyz'
alpha_upper='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
numeric='0123456789'
punc='.,!?~_@#$&|:;\\'
brackets='(){}[]'
math='+-/*=^%'
double_quote='"'
single_quote="'"
back_quote='`'
whitespace=' \t'
```

```python
idx_to_char=alpha_lower+alpha_upper+numeric+punc+brackets+math+double_quote+single_quote+back_quote+whitespace
print(idx_to_char)
char_to_idx={}
for i,v in enumerate(idx_to_char):
    char_to_idx[v]=i

# print(char_to_idx)
m=len(idx_to_char)
print(m)
```

```python
reflector_info_file='reflector_info.txt'
f=open(reflector_info_file,'w')
f.close()
arr=np.random.choice(m,size=m,replace=False)
# print(arr)
n=np.random.randint(0,int(m/4))
print(n)
ref=np.empty((1,m),dtype=np.int64)

for i in range(0,m,2):
    if m-i>n:
        ref[0][arr[i]]=arr[i+1]
        ref[0][arr[i+1]]=arr[i]
        print(str(arr[i])+' '+str(arr[i+1]))
    else:
        ref[0][arr[i]]=arr[i]
        if i+1<m:
            ref[0][arr[i+1]]=arr[i+1]

ref=np.reshape(ref,m)
ser=pd.Series(ref)
print(ser.to_string())

f=open(reflector_info_file,'a')
np.savetxt(f,ref,fmt='%1d')
f.close()
```

```python
arr=np.random.choice(m,size=m,replace=False)
n=np.random.randint(0,m)
plugboard=np.empty((1,m),dtype=np.int64)

for i in range(0,m,2):
    if m-i>n:
        plugboard[0][arr[i]]=arr[i+1]
        plugboard[0][arr[i+1]]=arr[i]
    else:
        plugboard[0][arr[i]]=arr[i]
        if i+1<m:
            plugboard[0][arr[i+1]]=arr[i+1]

```

```python
def construct_mapping(x,y,M):
    # returns numpy array containing [1,2,...,M] with y in x-th posn
    arr=np.arange(M)
    arr=(arr-(x-y)+M)%M
    return arr

def construct_rev(rotor):
    rev=np.empty_like(rotor)
    for i,v in enumerate(rotor):
        rev[v]=i
    return rev

class enigma:
    def __init__(self,m,rotor_info_file,selected_rotors,start_pos,reflector_info_file,plugboard):
        self.char_cnt=m
        # select 3 out of 8 rotors
        rotor_set=np.loadtxt(rotor_info_file,dtype=np.int64)
        self.rotors=[]
        self.rotors.append(rotor_set[selected_rotors[0]-1])
        self.rotors.append(rotor_set[selected_rotors[1]-1])
        self.rotors.append(rotor_set[selected_rotors[2]-1])
        # setup inverse rotor mappings
        self.rotors_inv=[]
        for rotor in self.rotors:
            self.rotors_inv.append(construct_rev(rotor))
        # align the rotors
        self.start_pos=start_pos
        self.reset()
        # set the reflector
        self.reflector=np.loadtxt(reflector_info_file,dtype=np.int64)
        # set the pugboard
        self.plugboard=plugboard
        
    def reset(self):
        # start_pos is a tuple of visible indices of the rotors during initial setting 
        self.zero_to_one=construct_mapping(self.start_pos[0],self.start_pos[1],self.char_cnt)
        self.one_to_two=construct_mapping(self.start_pos[1],self.start_pos[2],self.char_cnt)
        self.two_to_one=construct_mapping(self.start_pos[2],self.start_pos[1],self.char_cnt)
        self.one_to_zero=construct_mapping(self.start_pos[1],self.start_pos[0],self.char_cnt)
        self.vis=list(self.start_pos)
    
    def rotate_rotor_zero(self):
        self.zero_to_one=(self.zero_to_one-1+self.char_cnt)%self.char_cnt
        self.one_to_zero=(self.one_to_zero+1)%self.char_cnt
        self.vis[0]=(self.vis[0]+1)%self.char_cnt
    
    def rotate_rotor_one(self):
        self.one_to_zero=(self.one_to_zero-1+self.char_cnt)%self.char_cnt
        self.zero_to_one=(self.zero_to_one+1)%self.char_cnt
        self.one_to_two=(self.one_to_two-1+self.char_cnt)%self.char_cnt
        self.two_to_one=(self.two_to_one+1)%self.char_cnt
        self.vis[1]=(self.vis[1]+1)%self.char_cnt
        
    def rotate_rotor_two(self):
        self.two_to_one=(self.two_to_one-1+self.char_cnt)%self.char_cnt
        self.one_to_two=(self.one_to_two+1)%self.char_cnt
        self.vis[2]=(self.vis[2]+1)%self.char_cnt
    
    def convert(self,plain_txt):
        self.reset()
        cipher=''
        
        for char in plain_txt:
            if char not in char_to_idx:
                cipher+=char
                continue
            # print(char)
            p=char_to_idx[char]
            # print(p)
            p=plugboard[p]
            # print(p)
            p=self.rotors[0][p]
            # print(p)
            p=self.zero_to_one[p]
            # print(p)
            p=self.rotors[1][p]
            # print(p)
            p=self.one_to_two[p]
            # print(p)
            p=self.rotors[2][p]
            # print(p)
            p=self.reflector[p]
            # print(p)
            p=self.rotors_inv[2][p]
            # print(p)
            p=self.two_to_one[p]
            # print(p)
            p=self.rotors_inv[1][p]
            # print(p)
            p=self.one_to_zero[p]
            # print(p)
            p=self.rotors_inv[0][p]
            # print(p)
            p=plugboard[p]
            # print(p)
            
            e=idx_to_char[p]
            cipher+=e
            
            self.rotate_rotor_zero()
            if self.vis[0]==0:
                self.rotate_rotor_one()
                if self.vis[1]==0:
                    self.rotate_rotor_two()
            
        
        return cipher
```
